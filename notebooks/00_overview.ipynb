# DSA基礎（Python）全体像

このノートは **ロードマップとスコープ** の入口です。各 Phase は専用ノートで学習し、そこに書かれた **完了条件** を満たしたら次へ進みます。

## スコープ（固定）

- **主目的**: データ構造・アルゴリズムについて、**何を保持し、どの操作がどのコストで、どんな不変条件で成り立つか**を説明できること。
- **到達イメージ**: 代表的な DS/アルゴリズムを **選べる・比較できる・計算量を言葉にできる**。実装は理解を固めるためのオプション。
- **補助線**: 問題の言語化、AI 出力のレビュー（境界・計算量・不変条件）、小さな検証。
- **スコープ外（最初は切る）**: 競プロの細かいテクニック全集、高度な数論・幾何、大規模システム設計の全体。

## 各 Phase の完了条件（出口基準）

詳細は各 Phase ノート先頭にも記載。満たしたら次の Phase へ。

1. **構造の説明**: 核となる DS/アルゴリズムを 1 つ以上選び、図や具体例で「状態の変化」と「なぜその操作がそのコストか」を説明できる。
2. **計算量**: 主要操作の最悪（必要なら平均）を短文で言える。可能ならトレードオフを 1 組比較する。
3. **根拠の残し方**（いずれか）: 手動トレース、対照的な例、標準 API の期待挙動の列挙、**任意で**ミニ実装やテスト。
4. **一枚メモ**: 各 Phase ノート末尾の **「感想」セル**に、腹落ち・曖昧な点を書く。
