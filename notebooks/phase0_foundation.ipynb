## 理解の確認（最初の 1 題）

**問い**: 次の 3 つの操作について、(a) 計算量の見積り（最悪・ならしでよい）(b) なぜそう言えるか、を短く説明する。必要なら下のコードを実行して挙動を確認する。

1. `list` に対する `append(x)`（末尾に追加）
2. `list` に対する `insert(0, x)`（先頭に挿入）
3. `deque` に対する `appendleft(x)`（先頭に追加）

**ヒント**: Python の `list` は可変長配列として実装されており、末尾追加はならし（償却）で O(1) とみなせることが多いが、先頭への挿入は要素をずらすため高くつきやすい。`deque` は両端操作向けの構造。

（下に自分の答え用の Markdown セルを追加してよい）

```python
from collections import deque

n = 5_000

# 末尾追加: ならし（償却）で O(1) とみなせることが多い
lst_tail = []
for i in range(n):
    lst_tail.append(i)

# 先頭挿入: 毎回要素をずらすので O(k)（k は現在の長さ）→ 全体で O(n^2)
lst_head = []
for i in range(n):
    lst_head.insert(0, i)

# deque の先頭追加: 両端操作として O(1) を意図した利用
dq = deque()
for i in range(n):
    dq.appendleft(i)

len(lst_tail), len(lst_head), len(dq)
```
