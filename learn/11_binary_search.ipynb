## 段階コード 1: 最小実装


````python
def lower_bound(a,x):
    l,r=0,len(a)
    while l<r:
        m=(l+r)//2
        if a[m]<x: l=m+1
        else: r=m
    return l

print(lower_bound([1,3,3,5],3))
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
def exists(a,x):
    i=lower_bound(a,x)
    return i<len(a) and a[i]==x

print(exists([1,3,3,5],5))
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
def min_ok(lo,hi,ok):
    l,r=lo,hi
    while l<r:
        m=(l+r)//2
        if ok(m): r=m
        else: l=m+1
    return l

print(min_ok(0,100, lambda x:x*x>=30))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
