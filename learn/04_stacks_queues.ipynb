## 段階コード 1: 最小実装


````python
from collections import deque
q=deque([1,2,3])
q.append(4)
print(q.popleft())
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
stack=[]
for x in [1,2,3]:
    stack.append(x)
print(stack.pop())
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
from collections import deque
class JobQueue:
    def __init__(self): self.q=deque()
    def push(self,job): self.q.append(job)
    def pop(self): return self.q.popleft() if self.q else None

j=JobQueue(); j.push('A'); j.push('B'); print(j.pop(), j.pop())
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
