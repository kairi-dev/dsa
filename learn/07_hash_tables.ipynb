## 段階コード 3: 実務寄り実装


````python
def dedup(seq):
    seen=set(); out=[]
    for x in seq:
        if x in seen: continue
        seen.add(x); out.append(x)
    return out

print(dedup([1,2,1,3,2]))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
