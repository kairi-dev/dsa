## 段階コード 3: 実務寄り実装


````python
def select_projects(cost, value, budget):
    idx=sorted(range(len(cost)), key=lambda i:value[i]/cost[i], reverse=True)
    picked=[]; used=0
    for i in idx:
        if used+cost[i]<=budget:
            picked.append(i); used+=cost[i]
    return picked

print(select_projects([3,5,2,7],[8,12,5,14],8))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
