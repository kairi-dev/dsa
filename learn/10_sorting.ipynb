# ソートアルゴリズム

**Phase 5: アルゴリズム**

導入: この章では『順序づけで後段処理を効率化』を、初学者が説明できるレベルで学びます。


## 用語辞典（詳細）
- ソートアルゴリズムの中核概念: 順序づけで後段処理を効率化。
- 入力: アルゴリズムが受け取るデータ。
- 出力: 求めたい結果。
- 境界ケース: 空入力、最小入力、重複、未到達など。
- 時間計算量: 入力増加時の処理時間の増え方。
- 空間計算量: 追加メモリの増え方。


## 概念説明
これは何か: 順序づけで後段処理を効率化。
使いどころ: 実務の処理選定や性能改善で使う。
何が良いか: 問題に合う手法を選びやすくなる。
注意点: 入力特性・境界条件・更新頻度を無視すると逆効果になる。


## 計算量・式
比較ソート下限Ω(n log n)


## ASCII図
```text
[3,1,4,2] -> [1,2,3,4]
```


## 段階コード 1: 最小実装


````python
arr=[('b',2),('a',3),('a',1)]
print(sorted(arr, key=lambda x:(x[0],x[1])))
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
nums=[3,1,4,2]
print(sorted(nums))
print(sorted(nums, reverse=True))
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
records=[{'name':'a','score':80},{'name':'b','score':90},{'name':'c','score':90}]
print(sorted(records, key=lambda r:(-r['score'], r['name'])))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
