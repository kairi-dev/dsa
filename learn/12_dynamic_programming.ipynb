## 用語辞典（詳細）
- 動的計画法の中核概念: 部分問題の再利用。
- 入力: アルゴリズムが受け取るデータ。
- 出力: 求めたい結果。
- 境界ケース: 空入力、最小入力、重複、未到達など。
- 時間計算量: 入力増加時の処理時間の増え方。
- 空間計算量: 追加メモリの増え方。


## 概念説明
これは何か: 部分問題の再利用。
使いどころ: 実務の処理選定や性能改善で使う。
何が良いか: 問題に合う手法を選びやすくなる。
注意点: 入力特性・境界条件・更新頻度を無視すると逆効果になる。


## 計算量・式
dp[i]=dp[i-1]+dp[i-2]


## ASCII図
```text
dp0 -> dp1 -> dp2 -> ...
```


## 段階コード 1: 最小実装


````python
def fib(n):
    dp=[0]*(n+1)
    if n>=1: dp[1]=1
    for i in range(2,n+1):
        dp[i]=dp[i-1]+dp[i-2]
    return dp[n]

print(fib(10))
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
def fib_roll(n):
    a,b=0,1
    for _ in range(n):
        a,b=b,a+b
    return a

print(fib_roll(10))
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
def min_cost_stairs(cost):
    n=len(cost)
    dp=[0]*(n+1)
    for i in range(2,n+1):
        dp[i]=min(dp[i-1]+cost[i-1], dp[i-2]+cost[i-2])
    return dp[n]

print(min_cost_stairs([10,15,20]))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
