## 用語辞典（詳細）
- 連結リストの中核概念: ノード参照で要素をつなぐ構造。
- 入力: アルゴリズムが受け取るデータ。
- 出力: 求めたい結果。
- 境界ケース: 空入力、最小入力、重複、未到達など。
- 時間計算量: 入力増加時の処理時間の増え方。
- 空間計算量: 追加メモリの増え方。


## 概念説明
これは何か: ノード参照で要素をつなぐ構造。
使いどころ: 実務の処理選定や性能改善で使う。
何が良いか: 問題に合う手法を選びやすくなる。
注意点: 入力特性・境界条件・更新頻度を無視すると逆効果になる。


## 計算量・式
先頭挿入O(1), k番目アクセスO(k)


## ASCII図
```text
head -> [1] -> [2] -> [3] -> None
```


## 段階コード 1: 最小実装


````python
class Node:
    def __init__(self, v, nxt=None):
        self.v=v
        self.nxt=nxt

head=Node(1, Node(2, Node(3)))
print(head.nxt.v)
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
def push_front(head, x):
    # 先頭へO(1)で追加
    return Node(x, head)

h=Node(2,Node(3))
h=push_front(h,1)
print(h.v, h.nxt.v)
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
def remove_first(head, target):
    # 最初に見つかったtargetを削除
    dummy=Node(0,head)
    prev,cur=dummy,head
    while cur:
        if cur.v==target:
            prev.nxt=cur.nxt
            break
        prev,cur=cur,cur.nxt
    return dummy.nxt

nh=remove_first(Node(1,Node(2,Node(3))),2)
print(nh.v, nh.nxt.v)
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
