## 用語辞典（詳細）
- 最短経路・MSTの中核概念: 重み付き経路の最小コスト計算。
- 入力: アルゴリズムが受け取るデータ。
- 出力: 求めたい結果。
- 境界ケース: 空入力、最小入力、重複、未到達など。
- 時間計算量: 入力増加時の処理時間の増え方。
- 空間計算量: 追加メモリの増え方。


## 概念説明

## 段階コード 1: 最小実装


````python
import heapq
g={0:[(1,4),(2,1)],2:[(1,2)],1:[(3,1)],3:[]}
INF=10**9
d=[INF]*4; d[0]=0
pq=[(0,0)]
while pq:
    dist,u=heapq.heappop(pq)
    if dist!=d[u]:
        continue
    for v,w in g[u]:
        nd=dist+w
        if nd<d[v]:
            d[v]=nd
            heapq.heappush(pq,(nd,v))
print(d)
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。

## 段階コード 3: 実務寄り実装


````python
def shortest_path_cost(g,s,t):
    import heapq
    INF=10**9
    dist={k:INF for k in g}; dist[s]=0
    pq=[(0,s)]
    while pq:
        d,u=heapq.heappop(pq)
        if u==t: return d
        if d!=dist[u]: continue
        for v,w in g[u]:
            nd=d+w
            if nd<dist.get(v,INF):
                dist[v]=nd
                heapq.heappush(pq,(nd,v))
    return INF

print(shortest_path_cost(g,0,3))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
