## 段階コード 1: 最小実装


````python
state='idle'
for e in ['start','stop','start']:
    if state=='idle' and e=='start':
        state='run'
    elif state=='run' and e=='stop':
        state='idle'
print(state)
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
def step(state,event):
    table={('idle','start'):'run',('run','stop'):'idle'}
    return table.get((state,event), state)

s='idle'
for e in ['start','stop','start']:
    s=step(s,e)
print(s)
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
def simulate(events):
    s='idle'; history=[s]
    for e in events:
        if s=='idle' and e=='start': s='run'
        elif s=='run' and e=='stop': s='idle'
        history.append(s)
    return history

print(simulate(['start','noop','stop']))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
