## 用語辞典（詳細）
- 配列: 同じ種類の値を順番に格納する入れ物。連続したデータ処理に向く。
- ndarray: NumPyの配列型。数値計算を高速に行うために最適化されている。
- ベクトル化: for文を減らし、配列全体を一度に演算する書き方。
- ブロードキャスト: 形が合う範囲で自動的に配列サイズを合わせる仕組み。
- スライス: 必要な範囲だけを切り出す操作。
- dtype: 要素の型。float64/int32などで精度と速度が変わる。


## 概念説明
これは何か: 配列処理を高速・簡潔に書くための基礎。
使いどころ: センサ時系列、ログの数値集計、前処理。
何が良いか: 可読性と速度を両立しやすい。
注意点: 型変換やコピー発生に気づかないとメモリが増える。


## 計算量・式
sum = Σa_i, mean = (1/n)Σa_i, z = (x-mean)/std

## 段階コード 1: 最小実装


````python
import numpy as np

a = np.array([1, 2, 3, 4], dtype=np.float64)
# 全要素を2倍する
print(a * 2)
# 平均値を計算する
print(a.mean())
````

### コード1の説明
- まず正しさを確認する最小構成。
- デバッグ時は小入力で手計算と照合する。


## 段階コード 2: 改良実装


````python
import numpy as np

arr = np.arange(8, dtype=np.float64)
# 平均0, 分散1に近づける標準化
z = (arr - arr.mean()) / arr.std()
print(z.round(2))
````

### コード2の説明
- 主に性能または可読性を改善。
- どの条件で有利かを説明する。


## 段階コード 3: 実務寄り実装


````python
import numpy as np

def moving_avg(x, w=3):
    # 連続値を幅wで平滑化する
    x = np.asarray(x, dtype=np.float64)
    kernel = np.ones(w) / w
    return np.convolve(x, kernel, mode='valid')

print(moving_avg([1, 2, 3, 10, 3, 2, 1], 3))
````

### コード3の説明
- 現場の制約（更新頻度・ノイズ・運用）を反映。
- 採用判断を言語化できるようにする。
