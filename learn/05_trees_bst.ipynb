## 用語辞典（詳細）
- 木構造・二分探索木の中核概念: 階層構造で探索を効率化。
- 入力: アルゴリズムが受け取るデータ。
- 出力: 求めたい結果。
- 境界ケース: 空入力、最小入力、重複、未到達など。
- 時間計算量: 入力増加時の処理時間の増え方。
- 空間計算量: 追加メモリの増え方。


## 概念説明
これは何か: 階層構造で探索を効率化。
使いどころ: 実務の処理選定や性能改善で使う。
何が良いか: 問題に合う手法を選びやすくなる。
注意点: 入力特性・境界条件・更新頻度を無視すると逆効果になる。

## 段階コード 1: 最小実装


````python
class Node:
    def __init__(self,v,l=None,r=None):
        self.v=v; self.l=l; self.r=r
root=Node(8, Node(3, Node(1), Node(6)), Node(10))
print(root.l.r.v)
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
def inorder(n):
    if n is None: return []
    return inorder(n.l)+[n.v]+inorder(n.r)

print(inorder(root))
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
def bst_search(n,x):
    while n:
        if x==n.v: return True
        n=n.l if x<n.v else n.r
    return False

print(bst_search(root,6), bst_search(root,9))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
