# DSA教科書ガイド（説明強化版）

この`learn`フォルダは、初学者が**用語を説明できる**ことを最優先にした教材です。
コードは「読むだけ」ではなく、**何をしているかを言葉で説明できる**状態を目指します。


## この教材の読み方
1. まず「用語辞典」を読み、言葉の意味を押さえる
2. 次に「概念説明」で **これは何か / 使いどころ / 利点 / 注意点** を理解する
3. 「計算量と式」で速度・メモリの見積もりを比較する
4. 「ASCII図」で処理の流れを頭の中で再現する
5. コードは **コメントと解説セル** をセットで読み、説明できるか確認する


## 学習のゴール
- 用語を短く定義できる
- どの構造・アルゴリズムを選ぶべきか理由を言える
- 小さな入力で挙動確認し、実務向けに改良できる


## フェーズ構成（CLAUDE.md準拠）
