## 段階コード 3: 実務寄り実装


````python
def bandwidth_allocate(reqs, cap):
    reqs=sorted(reqs,key=lambda x:x[1]/x[0], reverse=True)
    use=[]; used=0
    for c,v in reqs:
        if used+c<=cap:
            use.append((c,v)); used+=c
    return use

print(bandwidth_allocate([(3,5),(5,7),(2,4)],7))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
