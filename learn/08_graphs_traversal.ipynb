## 用語辞典（詳細）

## 段階コード 1: 最小実装


````python
from collections import deque
g={0:[1,2],1:[3],2:[3],3:[]}
q=deque([0]); seen={0}
while q:
    u=q.popleft()
    for v in g[u]:
        if v not in seen:
            seen.add(v); q.append(v)
print(seen)
````

### コード1の説明
- まず最小の動く例で、処理手順を追えるようにする。
- どの入力でどう出力が変わるかを確認する。


## 段階コード 2: 改良実装


````python
def dfs(u,g,seen):
    seen.add(u)
    for v in g[u]:
        if v not in seen:
            dfs(v,g,seen)

seen=set(); dfs(0,g,seen); print(seen)
````

### コード2の説明
- 可読性・性能・再利用性のどれを改善したかを明示する。
- 最小実装との差分を言葉で説明できることが目標。


## 段階コード 3: 実務寄り実装


````python
def reachable(g,s,t):
    from collections import deque
    q=deque([s]); seen={s}
    while q:
        u=q.popleft()
        if u==t: return True
        for v in g.get(u,[]):
            if v not in seen:
                seen.add(v); q.append(v)
    return False

print(reachable(g,0,3))
````

### コード3の説明
- 実運用で必要になる条件分岐・制約・保守性を反映。
- 採用理由を『速度・メモリ・実装コスト』で説明できるようにする。
